## Defining the state to pass messages as tokens

```python
try:
    from typing import TypedDict, Annotated
except ImportError:
    from typing_extensions import TypedDict, Annotated
import operator
from langchain_core.messages import AnyMessage,HumanMessage,AIMessage
class GraphState(TypedDict): #Any name instead of TypedDict will work
    messages: Annotated[list[AnyMessage], operator.add] ## Operator.add is used to concatenate the list of messages when we update the state with new messages.
```
